```python
# https://www.kaggle.com/jsphyg/weather-dataset-rattle-package 

# import data 
data = pd.read_csv('weatherAUS.csv')

# drop 'risk' column according to dataset creator
data.drop(columns=['RISK_MM'], inplace=True)

data
```

### Data exploration

```python
# what columns ?
data.columns
```

```python
# drop some data, juged to be not too important
target = data[['RainTomorrow']]

data = data[['Date', 'Location', 'MinTemp', 'MaxTemp', 'Evaporation',
       'WindGustDir', 'WindGustSpeed', 'WindDir9am', 'WindDir3pm',
       'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm',
       'Pressure9am', 'Pressure3pm', 'Temp9am',
       'Temp3pm', 'RainToday']]
```

```python
data.Location.value_counts() # filter only one city, give one data line per day ?
```

```python
# example for one date with multiple data line : => only one date if we filter on one location
z = data.Date.value_counts().index[0]
data['Location'].loc[data.Date == z].value_counts()

# we drop date
data.drop(columns=['Date'], inplace=True)
```

```python
# reduce data, only location = Sydney
target = target.loc[data.Location == 'Sydney'].reset_index(drop=True)
data = data.loc[data.Location == 'Sydney'].reset_index(drop=True)

# drop location column
data.drop(columns=['Location'], inplace=True)
```

```python
# min/max temp feature

x = data['MinTemp']
y = data['MaxTemp']
z = target['RainTomorrow']
plt.plot(x,z, 'r+')

plt.show()

plt.plot(y,z, 'b+',)
plt.show()

# compute temperature delta :
data['SquareDeltaTemperature'] = (x - y)**2
plt.plot(data['SquareDeltaTemperature'],z, 'g+')
plt.show()


# drop min/max temperature columns
data.drop(columns=['MinTemp','MaxTemp'], inplace=True)
```

```python
# wind direction feature 'WindGustDir', 'WindDir9am', 'WindDir3pm'

# direction => probably not very useful (same goes for 3pm and 9 am)
x = data['WindGustDir']
print(x)
print(x.value_counts())


# we drop these columns
data.drop(columns = ['WindGustDir', 'WindDir9am', 'WindDir3pm'], inplace=True)
```

```python
# wind speed :  'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm'

# speed
x = data['WindGustSpeed']
print(x)
print(x.describe())

plt.plot(x, target['RainTomorrow'],'b+')
plt.show()

# 9 am speed
plt.plot(data['WindSpeed3pm'], target['RainTomorrow'], 'g+')
plt.show()

# 3 pm speed 
plt.plot(data['WindSpeed3pm'], target['RainTomorrow'], 'r+')
plt.show()


# keep only Gust speed, since the most relevant (max gust)
data.drop(columns = ['WindSpeed9am', 'WindSpeed3pm'], inplace=True)
```

```python
# humidity feature 'Humidity9am', 'Humidity3pm'

data['HumidityDelta'] = data['Humidity3pm'] - data['Humidity9am']

# humidity delta seems more relevant
plt.plot(data['HumidityDelta'],  target['RainTomorrow'], 'r+')

# drop humidity columns
data.drop(columns=['Humidity3pm', 'Humidity9am'], inplace = True)
```

```python
# pressure feature  'Pressure9am', 'Pressure3pm'

data['PressureDelta'] = data['Pressure3pm'] - data['Pressure9am']

plt.plot(data['PressureDelta'], target['RainTomorrow'], 'g+')

# delta seems more relevant => drop original pressure columns
data.drop(columns = ['Pressure9am','Pressure3pm'], inplace =True)
```

```python
# temperature feature  'Temp9am', 'Temp3pm'

x = data['Temp3pm'] - data['Temp9am']

plt.plot(x, target['RainTomorrow'], 'r+')

# min/max contains probably more useful information => we drop these columns
data.drop(columns=['Temp3pm','Temp9am'], inplace = True)
```

### Feature Engineering

```python
# number of days without rain before


# cant compute for the firt rain day
x =  data['RainToday']

rain_idx = np.where(x == 'Yes')[0]

start_idx = rain_idx.min()+1

liste = []

for index, elem in data.iterrows():
    
    if index >= start_idx  :

        day_count = index - rain_idx[np.where(rain_idx < index)[0].max()] -1
        
        liste.append(day_count)
        
    else: 
        liste.append(np.nan)

        
data['DayWithoutRain'] = liste


# drop data until the day after the first rainy day
data.drop(index=[i for i in range(start_idx)], inplace= True)
target.drop(index=[i for i in range(start_idx)], inplace= True)
```

```python
# plot this new feature

plt.plot(data['DayWithoutRain'], target['RainTomorrow'], 'r+')
```

```python
# transform non numerical features into numerical & deal with 'nan' values

target.replace({'Yes':1,'No':0}, inplace =True)

data.replace({'Yes':1,'No':0}, inplace =True)

# evaporation, windgustspeed : nan => 0
data.replace({'WindGustSpeed':np.nan , 'Evaporation' : np.nan}, 0, inplace=True)
```
